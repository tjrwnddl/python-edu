```python
class FourCal:
    def __init__(self, first, second): # 생성자:멤버필드를 초기화 하기 위해서
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
    def div(self):
        self.result = self.first / self.second  
```

```python
class MoreFourCal(FourCal):
    def sub(self):
        self.result = self.first - self.second 
    def mul(self):
        self.result = self.first * self.second
    def div(self): # 재정의 : override
        if self.second == 0:
            self.result = 0
        else:
            self.result = self.first / self.second
```

```python
cal8 = MoreFourCal(20, 0)
cal8.div()
print(cal8.result)
# 클래스의 구성 요소 : 멤버필드(멤버변수), 메서드, 생성자
# 메서드에는 self
# 멤버필드를 초기화 : 생성자
# 상속 : 다른 클래스에 정의된 걸 가지고 온는것
# 상속받은 메서드를 다시 정의 : override
```

```python
import requests  # 스크래핑
from bs4 import BeautifulSoup # 크롤링
import pandas # 데이터분석
import matplotlib as mpl 
import matplotlib.pyplot as plt # 시각화
url = "https://www.weather.go.kr/w/obs-climate/land/city-obs.do"
source = requests.get(url)
soup = BeautifulSoup(source.content, "html.parser")
#pint(soup)
table = soup.find("table",{'class':'table-col'})

data = []

for tr in table.find_all('tr'):
    tds = list(tr.find_all("td"))
    for td in tds:
        if td.find('a'):
            point = td.find('a').text 
            temp =  tds[5].string
            humidity = tds[9].string
            data.append([point,temp,humidity])

#데이터 처리
with open('weather.csv','w', encoding='utf-8') as f:
    f.write('지역, 온도, 습도\n')
    for i in data:
        f.write("{0},{1},{2}\n".format(i[0],i[1],i[2]))

# 분석할 데이터 저장
df = pandas.read_csv("weather.csv", index_col ="지역", encoding="utf-8")
df
city_df = df.loc[['서울','인천','대전','대구','광주','부산','울산']]
city_df

font_name = mpl.font_manager.FontProperties(fname="C:\Windows\Fonts\malgun.ttf").get_name()
mpl.rc('font', family=font_name)

ax = city_df.plot(kind="bar", title="날씨", figsize=(12,4))
plt.show()
```

```python
# 2차원 배열 붙이기
x2 = np.array([1, 2, 3])   # 1차원 배열
grid = np.array([[9, 8, 7], 
                 [6, 5, 4]]) # 2차원 배열
grid1 = np.array([[9, 8, 7], 
                 [6, 5, 4]]) # 2차원 배열
print(x2)
print(grid)
print(grid1)
# z = np.concatenate((x2, grid), axis=1)
# 차원이 다르면 concatenate함수를 사용할 수 없다.
z = np.concatenate((grid, grid1)) # axis=0은 기본설정
print(z)
z = np.concatenate((grid, grid1), axis=0) # x축
print(z)
z = np.concatenate((grid, grid1), axis=1) # y축
print(z)

# 차원이 다른 경우 : vstack(열의 차원이 일치), hstack(행의 차원이 일치)
z = np.vstack((x2 , grid))
print(z)
grid3 = np.array([[1,2],[3,4]])
print(grid3)
z = np.hstack((grid, grid3))
print(z)
```
